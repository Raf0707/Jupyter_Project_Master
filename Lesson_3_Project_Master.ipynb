```python
df_l.groupby('client').agg({'success': 'sum'}).query('success == success.max()').sort_values('client').index.tolist()
```

```python
df_l.query('success == 1').groupby('platform', as_index=False).agg({'success': 'count'}).sort_values(by='success', ascending=False)
```

```python
df_l.merge(df_ud, how='left', on='client').query('premium == 1').groupby('platform', as_index=False).agg({'success': 'count'}).sort_values(by='success', ascending=False)
```

```python
df_ul = df_ud.merge(df_l)
```

```python
df_ul
```

```python
true = df_ul.query('premium == True')
false = df_ul.query('premium == False')
sns.distplot(false.age)
sns.distplot(true.age)
```

```python
df_ul_1 = df_ul.query('success == True').groupby('client', as_index=False).agg({'success': 'count'})

df_def_kxf = df_ul_1.groupby('success', as_index= False).agg({'client': 'count'})

sns.barplot(x='success', y='client', data=df_def_kxf, color='pink', alpha=1)
```

```python
plt.figure(figsize=(18, 8))
sns.countplot(x="age", data=df_l.merge(df_ud, how='left', on='client').query("success == True & platform == 'computer'"))
plt.show()
```
