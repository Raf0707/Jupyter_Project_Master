```python
def Samaya_moya_moya(S):
    if S.find('Mr.') >= 0:
        return
    elif S.find('Miss. ') >= 0:
        ss = S.find('Miss. ')
        S = S[ss + 6:]
        cc = S.find(' ')
        
        return S[:cc]
        
    else:
        rs = S.find('(')


        S = S[rs + 1:]
        sr = S.find(' ')
        d = S[:sr]
        if d[-1] == ',':
            return d[:-1]
        else:
            return d
```

```python
df.Name = df.Name.apply(Samaya_moya_moya)
```

```python
df.Name = df.Name.dropna()
```
